##2018

```python
# Obtener todos los valores únicos de la columna "tipo_pedido"
tipoPedido_unicos = SAC['tipo_pedido'].unique()

# Top 10 de los barrios más frecuentes del 2018
top_barrio18 = sac18['nombre_barrio'].value_counts()[:10].index

# Filtrar los datos para incluir solo los barrios del top 10
barrios_top10_18 = sac18[sac18['nombre_barrio'].isin(top_barrio18)]

# Crear un DataFrame vacío para almacenar los resultados
df_top_barrios18_pedidos = pd.DataFrame({'Barrio': top_barrio18})

# Iterar a través de los tipos de pedido y acumular los datos en el DataFrame
for tipo_pedido in tipoPedido_unicos:
    # Filtrar los datos por tipo de pedido
    datos_tipo_pedido = barrios_top10_18[barrios_top10_18['tipo_pedido'] == tipo_pedido]

    # Calcular la frecuencia de los barrios y almacenar los resultados en un DataFrame
    frecuencia_por_barrio = datos_tipo_pedido['nombre_barrio'].value_counts().reset_index()
    frecuencia_por_barrio.columns = ['Barrio', tipo_pedido]

    # Combinar los resultados con el DataFrame principal
    df_top_barrios18_pedidos = pd.merge(df_top_barrios18_pedidos, frecuencia_por_barrio, on='Barrio', how='left')

# Llenar valores faltantes con cero
df_top_barrios18_pedidos.fillna(0, inplace=True)

# Convertir las columnas de tipos de pedido a enteros
for tipo_pedido in tipoPedido_unicos:
    df_top_barrios18_pedidos[tipo_pedido] = df_top_barrios18_pedidos[tipo_pedido].astype(int)

# Imprimir el resultado final
df_top_barrios18_pedidos.to_csv('/content/drive/MyDrive/Cumbre Datos/top_barrios_2018.csv', index=False)
df_top_barrios18_pedidos
```

```python
# Crear un DataFrame con los datos
data = {
    'Año': [2018, 2019, 2020, 2021, 2022, 2023],
    'Reposición de focos': [6891, 6799, 7876, 8318, 8561, 7014],
    'Ruidos molestos': [2861, 3543, 2812, 2826, 3237, 2029],
    'Zonas apagadas': [2278, 2748, 2658, 3143, 2977, 2490]
}

df = pd.DataFrame(data)

# Función para calcular la variación absoluta y porcentual
def calcular_variacion(df, columna, año_inicial, año_final):
    valor_inicial = df[df['Año'] == año_inicial][columna].values[0]
    valor_final = df[df['Año'] == año_final][columna].values[0]
    variacion_absoluta = valor_final - valor_inicial
    variacion_porcentual = ((valor_final - valor_inicial) / valor_inicial) * 100
    return variacion_absoluta, variacion_porcentual

# Calcular la variación entre 2018 y 2020
variacion_2018_2020 = {}
for columna in ["Reposición de focos", "Ruidos molestos", "Zonas apagadas"]:
    variacion_absoluta, variacion_porcentual = calcular_variacion(df, columna, 2018, 2020)
    variacion_2018_2020[columna] = {
        'Variación Absoluta': variacion_absoluta,
        'Variación Porcentual': variacion_porcentual
    }

# Calcular la variación entre 2020 y 2023
variacion_2020_2023 = {}
for columna in ["Reposición de focos", "Ruidos molestos", "Zonas apagadas"]:
    variacion_absoluta, variacion_porcentual = calcular_variacion(df, columna, 2020, 2023)
    variacion_2020_2023[columna] = {
        'Variación Absoluta': variacion_absoluta,
        'Variación Porcentual': variacion_porcentual
    }

# Imprimir los resultados
print("Variación entre 2018 y 2020:")
print(pd.DataFrame(variacion_2018_2020).T)

print("\nVariación entre 2020 y 2023:")
print(pd.DataFrame(variacion_2020_2023).T)
```
